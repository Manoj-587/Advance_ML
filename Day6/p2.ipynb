```python
# =========================================================
# Supervised Learning  Initial Data Inspection
# Outlier Boxplot & Correlation Heatmap
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# 1. Read dataset filename from user
# ---------------------------------------------------------


# ---------------------------------------------------------
# 2. Load dataset
# ---------------------------------------------------------
df = pd.read_csv("ML470_S6_Mobile_Data_Practice.csv")

# ---------------------------------------------------------
# 3. Separate input features and target
# ---------------------------------------------------------
X = df.drop("price_range", axis=1)

# ---------------------------------------------------------
# 4. Outlier Assessment using Boxplots
# ---------------------------------------------------------
plt.figure(figsize=(14, 5))
sns.boxplot(data=X)
plt.xticks(rotation=45)
plt.title("Outlier Assessment Using Boxplots")
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 5. Feature Correlation Analysis using Heatmap
# ---------------------------------------------------------
corr_matrix = X.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="YlGnBu",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Feature Correlation Analysis Using Heatmap")
plt.tight_layout()
plt.show()
```
