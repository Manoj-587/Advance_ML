```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Read the Iris dataset from CSV
df = pd.read_csv("Iris(in).csv")


# 2. Extract feature columns
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# 3. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Compute hierarchical clustering using Ward’s method
Z = linkage(X_scaled, method='ward')

# 5. Plot dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram (Ward’s Method)")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.tight_layout()
plt.show()

# 6. Extract cluster labels (example: 3 clusters)
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = hc.fit_predict(X_scaled)

# 7. Print cluster labels
print("Cluster Labels:")
print(cluster_labels)
```
